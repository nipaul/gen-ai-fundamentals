```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch
import sentencepiece as spm
import pprint

model_name = "t5-small"

# Initialize tokenizer and model
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Input text
text = "Home"

# Tokenize input text, extract embeddings and print both input IDs and embeddings
input_ids = tokenizer.encode(text, return_tensors="pt", max_length=512, truncation=True)
with torch.no_grad(): token_embeddings = model.shared(input_ids)    

pprint.PrettyPrinter(indent=4).pprint({
    "Input IDs": input_ids, "Token Embeddings": token_embeddings
})
```

Snippet with AutoTokenizer and BERT model

```python
from transformers import AutoTokenizer, AutoModel
import torch
import sentencepiece as spm
import pprint

model_name = "bert-base-uncased"

# Initialize tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Input text
text = "Home"

# Tokenize input text, extract embeddings and print both input IDs and embeddings
input_ids = tokenizer.encode(text, return_tensors="pt", max_length=512, truncation=True)
with torch.no_grad(): token_embeddings = model(input_ids)    

pprint.PrettyPrinter(indent=4).pprint({
    "Input IDs": input_ids, "Token Embeddings": token_embeddings
})
```
